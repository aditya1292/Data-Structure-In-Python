# Queue 

#### Queue is a Linear Data Structure, It stores data in  "First In First Out" FIFO manner item which is inserted first is removed first


### Basic Operations on Queue 

### Using List Implementation 


```python
Q=[] #Create an Empty List 
rear=0 #rear size is set to Zero 
size=int(input("please enter the size of the Queue"))  #Input for the size of the queue
```

## Operations on Queue 
#### 1. ENQUEUE
            Enqueue : Adding an Element to queue is called enqueue   
#### 2. DEQUEUE
            Dequeue : Removing an Element form queue is called dequeue 

### Overflow
        Overflow is the condition when the queue is completely full but we try to add more element in it.  
### Underflow
        Underflow is the condition when the queue is completly empty and we try to remove more element from it.

```python
print("1.EnQueue\n2.Dequeue\n3.Display\n4.Exit")
choice=int (input("Please enter your choice "))
if choice==1:
  if rear==size:  #Condition of Stack Overflow
    print("Over Flow")
  else:
    data=int (input("Please enter your data for EnQueue " " "))
  Q.append(data) #.append is used to enqueue a data in a queue
  print(data," Enqueued " "")
elif choice==2:
  if Q==[]: #Condition of Stack Underflow 
    print("Under flow")
  else:
    data=Q.pop(0) # pop is used to remove(Dequeue) an element from queue 
    print(data,"Dequeued ")
elif choice==3:
  if Q==[]:
    print("Queue is Empty ")
  for T in Q:
    print(T,"<--",end=" ")
 
  print("*"*20,"\n\n\n") # '\n'is the newline character used to create new line 
elif choice==4:
  pass
else:
    print("Invalid option ")
```
